```r

# Visualizar informações sobre as estações meteorológicas disponíveis
estacoes <- BrazilMet::see_stations_info()
estacoes
```

```r

# Baixar dados diários de uma estação específica (A521) para o ano de 2020
df <- download_AWS_INMET_daily(
  stations = estacoes$station_code,
  start_date = "2019-01-01",
  end_date = "2023-12-31"
)

# Visualizar os dados baixados
View(df)
```
